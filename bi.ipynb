```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
```

```python
with sns.axes_style('whitegrid'):
    Grafico = sns.lineplot(data = flights , x = 'month', y = 'passengers', hue = 'year')
```

# Granulidade media

# Min = 1949

# Max = 1960

```python
month = {'Jan': 1,
         'Feb': 2,
         'Mar': 3,
         'Apr': 4,
         'May': 5,
         'Jun': 6,
         'Jul': 7,
         'Aug': 8,
         'Sep': 9,
         'Oct': 10,
         'Nov': 11,
         'Dec': 12}
```

```python
flights['month'] = flights['month'].apply(lambda x : month[x])
```

```python
flights['year-month'] = flights['year'].astype(str) + '-' + flights['month'].astype(str).str.zfill(2)
```

```python
flights['year-month'] = pd.to_datetime(flights['year-month'], format='%Y-%m').dt.strftime('%Y-%m-%d')
```

```python
flights = flights[['year-month','year','month','passengers']]
```

```python
flights.to_csv('flights.csv',index = False)
```
