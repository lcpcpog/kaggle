# Parental level of education:

```python
best_foundation ={
    "some high school": 0,
    "high school" : 1,
    'some college' : 2,
    "associate's degree": 3,
    "bachelor's degree" : 4,
    "master's degree" : 5
}
df['parental level of education correletion'] = df['parental level of education'].apply(lambda valor : best_foundation[valor])
```

```python
correlation_math = np.corrcoef(df['parental level of education correletion'], df['math score'])
correlation_reading = np.corrcoef(df['parental level of education correletion'], df['reading score'])
correlation_writing = np.corrcoef(df['parental level of education correletion'], df['writing score'])
correlation_media = np.corrcoef(df['parental level of education correletion'], df['media score'])

print(f'correlation math: {correlation_math}')
print()
print(f'correlation reading: {correlation_reading}')
print()
print(f'correlation writing: {correlation_writing}')
print()
print(f'correlation media: {correlation_media}')
```
